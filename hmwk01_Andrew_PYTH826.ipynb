### P1.PY




```
"""
Typcasting w. Integers & Floats
"""

# Convert these numbers into floats and back. Print out each result as well as its data type.



five = 5
zero = 0
neg_8 = -8
neg_22 = -22

float_five = float(five)
float_zero = float(zero)
float_neg_8 = float(neg_8)
float_neg_22 = float(neg_22)

# Varibables are converted to floats and stored as new variables

print(float_five, float_zero, float_neg_8, float_neg_22)
print(type(float_five), type(float_zero), type(float_neg_8), type(float_neg_22))


# Printing new variable types using 'type' and 'isinstance' functions

x_float_five = int(float_five)
x_float_zero = int(float_zero)
x_float_neg_8 = int(float_neg_8)
x_float_neg_22 = int(float_neg_22)

# Variable are convertes back to intergers and stored again as new variables

print(x_float_five, x_float_zero, x_float_neg_8, x_float_neg_22)
print(type(x_float_five), type(x_float_zero), type(x_float_neg_8), type(x_float_neg_22))

# Printing new variable types using 'type' and 'isinstance' functions

print(isinstance(float_zero, float))
print(isinstance(float_zero, int))
print(isinstance(float_zero, str))

print(isinstance(x_float_zero, float))
print(isinstance(x_float_zero, int))
print(isinstance(x_float_zero, str))

# Using 'isinstance' function to test/identify various data types
```

### P1.PY




```
"""
Integer & Float Operators
"""

# Complete the specified math operations. Do the next operation on the result from the previous operation.


orig_var = 100

# Add 50

num1 = orig_var + 50

# Subtract 90

num2 = num1 - 90

# Multiply 10

num3 = num2 * 10

# Divide 150

num4 = num3 / 150

# Modulus 3

num5 = num4 % 3
print(num5) #1.0


# or using assignment operators

orig_var = 100

orig_var += 50 #150

orig_var -= 90 #60

orig_var *= 10 #600

orig_var /= 150 # 4.0

orig_var %= 3 # 1.0

print(orig_var)


```
